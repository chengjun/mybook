````python
import pylab as plt
plt.style.use('seaborn')
plt.rcParams['font.sans-serif'] = ['SimHei']  # Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫‰∏≠ÊñáÊ†áÁ≠æ
plt.rcParams['axes.unicode_minus'] = False  # Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫Ë¥üÂè∑

x = range(100)
y = [i**2 for i in x] 
plt.plot(x, y, 'r--', label = '‰∫∫ÂïäÔºåËÆ§ËØÜ‰Ω†Ëá™Â∑±ÔºÅ')
plt.xlabel('Êàë‰∏ÄÊó†ÊâÄÁü•ÔºÅ', fontsize = 16)
plt.ylabel('Êú™ÁªèÁúÅÂØüÁöÑ‰∫∫Áîü‰∏çÂÄºÂæóÂ∫¶Ëøá„ÄÇ', fontsize = 16)
plt.title('ËãèÊ†ºÊãâÂ∫ï', fontsize = 20)
plt.legend(fontsize = 16)
plt.show()
````
